```python
recent_grads['Sample_size'].hist(bins=30, range=(0, 3000))
```

The most common sample size by a high margin is between 0 and 100, representing about 45% of the dataset.  About 80% of the whole dataset consists of sample sizes of under 500.

The most common median income is roughly 30-35k by a substantial margin and is almost 1/3rd of the total dataset.  The runner up in frequency, 40-45k, occurs less than half as frequently.  The distribution is right skewed.

```python
recent_grads['ShareWomen'].hist(bins=20)
```

The distribution of proportion of women in these college majors in this dataset is left skewed, with the most common proportions between 0.6 and 0.8.

```python
recent_grads['Unemployment_rate'].hist(bins=20)
```

The unemployment rate in this dataset is close to a normal distribution, with a long tail to the right.  The most common rates are roughly 0.04 - 0.10.

```python
recent_grads['Women'].hist(range=(0, 60000))
```

Women seem to be be more highly represented in the majors with a larger sample size than men.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot.hexbin(x=['Sample_size'], y=['Unemployment_rate'], gridsize=30, figsize = (8, 6), colormap='gnuplot', title = "American College Grads: Sample Size vs Unemployment Rate")
```

The majors with the largest sample sizes are roughly in the middle of the distribution in terms of unemployment rates.  The smallest programs represent outliers in both directions in terms of unemployment rates. The hexpot shows that the highest concentration of below average unemployment rates is in the majors with the smallest sample sizes.

```python
recent_grads.plot('Full_time', 'Median', 'scatter')
```

```python
recent_grads.plot.hexbin(x=['Full_time'], y=['Median'], gridsize=30, figsize = (8, 6), colormap='gnuplot', title = "American College Grads: Full Time Workers vs Median Income")
```

The majors with the highest numbers of grads working full time are on the low to middle range of the income distribution.  The two programs with roughly 250,000 full time workers both had median incomes of 40,000 or less.  All of the programs with median grad income of 60,000 or more had 75,000 or less few time workers.

```python
recent_grads.plot('ShareWomen', 'Unemployment_rate', 'scatter')
```

There appears to be no correlation between the proportion of grads for a major who are women, and the unemployment rate for the major.

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
recent_grads.plot.hexbin(x=['Sample_size'], y=['Median'], gridsize=30, figsize = (8, 6), colormap='gnuplot', title = "American College Grads: Sample Size vs Median Income")
```

Most of the data points in this dataset are for sample sizes of less than 1000.  The smaller sample sizes represent median incomes throughout the range.  As sample sizes increase past 1000, there are no data points showing median incomes above roughly 50000 for median income.  The higher incomes seem to be concentrated among the less common college majors.

```python
recent_grads.plot('Men', 'Median', 'scatter')
```

```python
recent_grads.plot.hexbin(x=['Men'], y=['Median'], gridsize=30, figsize = (8, 6), colormap='gnuplot', title = "American College Grads: Number of Male Grads vs Median Income")
```

The distribution of men vs median income looks very similar to that of the total sample size vs income.  No additional trend relating to gender and median income is apparent here.

```python
recent_grads.plot('Women', 'Median', 'scatter')
```

```python
recent_grads.plot.hexbin(x=['Women'], y=['Median'], gridsize=30, figsize = (8, 6), colormap='gnuplot', title = "American College Grads: Number of Female Grads vs Median Income")
```

Visualizing the median incomes by number of women in a major, it becomes apparent that compared to the number of men by major, the majors completed by more women have a lower median income.  The only datapoint with a median income of 60,000 or greater has about 25000 female graduates.  Conversely, majors by men with median incomes of around 60,000 occur for programs with 75,000 male graduates.  It would seem that although the unemployment rate is unaffected by the number of female grads, the gender is at least somewhat correlated median income.

```python
recent_grads.plot('ShareWomen', 'Median', 'scatter')
```

This chart shows a slight negative correlation between proportion of women in a college major and median income, reinforcing the earlier finding.

```python
scatter_matrix(recent_grads[['ShareWomen', 'Median']])
```

There it is very stark that an increased proportion of women in a college major is inversely proportionate to median income.

```python
scatter_matrix(recent_grads[['ShareWomen', 'Low_wage_jobs']])
```

The college majors with the most low-wage service jobs are also those with the highest proportion of women.
